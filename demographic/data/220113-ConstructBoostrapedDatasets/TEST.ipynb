```python
import numpy as np
import moments
import os
import pandas as pd
import gzip
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_spectrum(file, compresed=True):
    """Load the SFS"""
    print(f'loading {file} ...')
    if compresed:
        with gzip.open(file, 'rb') as f:
            return pickle.load(f)
    else:
        with open(file, 'rb') as f:
            return pickle.load(f)
        
sfs = load_spectrum('data/whole-genome/spectrum-cat_synonymous.pkl.gz')
```
