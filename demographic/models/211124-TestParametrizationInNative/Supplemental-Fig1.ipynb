```python
import demes
import demesdraw
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (3, 3)


# define the color for plotting
# The idea of this notebook is just to
#Â make nicer figures

COLORS = dict(
    ancestral='#453581',
    AMH='#98d83e',
    YRI='#fde725',
    IBS='#5ec962',
    MXB='#3b528b',
    CHB='#21918c'
)

def draw_model(mdl, positions):
    """
    Args:
        mdl: demes-graph
        positions
    """
    demesdraw.tubes(
        mdl,
        positions=positions,
        log_time=True,
        colours=COLORS,
        num_lines_per_migration=1,
    )
```

```python
mdl_1 = demes.load('results/01ConstantSize-best.yml')

w = demesdraw.utils.separation_heuristic(mdl_1)
positions = dict(ancestral=0, AMH=0, YRI=0, OOA=0.7 * w, IBS=0.6 * w, CHB=1.1 * w, MXB=1.4 * w)

draw_model(mdl_1, positions)
plt.savefig('plots/01ConstantSize-demes-fancy.pdf')
```
