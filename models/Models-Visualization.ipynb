# Inferred demographic models

```python
import demes
import demesdraw
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['figure.dpi'] = 300

# colors for plots
COLORS = dict(
    ancestral='#453581',
    AMH='#98d83e',
    YRI='#fde725',
    IBS='#5ec962',
    MXB='#3b528b',
    CHB='#21918c'
)
```

## Model 1 `m1-out-of-africa.yml`: Four populations out of Africa

```python
m1 = demes.load('m1-out-of-africa.yml')
demesdraw.tubes(
    m1,
    log_time=True,
    colours=COLORS,
    num_lines_per_migration=1,
);
```

## Model 2 `m2-Mexico-admixture.yml`: Admixture in Mexico

```python
COLORS_ = COLORS.copy()
COLORS_['MXL'] = '#440154'
m2 = demes.load('m2-Mexico-admixture.yml')

demesdraw.tubes(
    m2,
    log_time=True,
    colours=COLORS_,
    num_lines_per_migration=1,
);
```
